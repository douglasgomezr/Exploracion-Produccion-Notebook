```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import cm
```

```python
############################
#Dominio temporal
############################
dt = 0.001
t = np.arange(0, 1, dt)
nt = t.size
f = 8
############################
#Frecuencua de Nyquist
############################
fn = 1/(2*dt)
```

```python
print('Frecuencia Nyquist temporal =', fn, '[Hz]')
print ('Frecuencia circular =', f, '[Hz]')
```

```python
signal = np.sin((2*3.1416)*f*t)
```

```python
def DFT1(x):
    'DFT a partir del producto interno'
    N = x.size
    n = np.arange(N)
    k = n.reshape((N,1))
    e = np.exp(-2j*np.pi*k*n/N)
    return np.dot(e,x)
```

```python
def DFT2(x):
    'DFT a partir del producto interno'
    N = x.size
    n = np.arange(N)
    k = n.reshape((N,1))
    e = np.exp(-1j*k*n/N)
    return np.dot(e,x)
```

```python
#Seal
sx = np.cos(-2*3.1416*f*t)
#DFT de la seal
sX1 = DFT1(sx)
sX2 = DFT2(sx)
```

```python
plt.figure(figsize=(8,3)) 

_ = plt.plot(abs(sX1))
_ = plt.xlabel('$k= 0,\ldots, N-1$', fontsize = 11, weight = 'semibold')
_ = plt.ylabel('Amplitud', fontsize = 11, weight = 'semibold')
_ = plt.grid(True, alpha = 0.5, color = 'grey', linestyle = ':')
```
