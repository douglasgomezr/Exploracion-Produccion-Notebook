## Ajuste de longitud de dos series de tiempo

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Definimos una serie de tiempo que emula datos en la escala de registros geof√≠sicos

```python
prof = np.arange(0, 2000, 0.1524) #Intervalo de muestreo en profundidad, de 0 a 2000 metros con un intervalo de muestreo de 0.1524 metros.
dtco = np.cos(0.3*prof) #Pseudo registro DTCO
```

```python
print('Longitud dtco:', len(dtco), 'muestras')
```

```python
plt.figure(figsize=[13,2])
_ = plt.plot(prof, dtco, 'r-')
_ = plt.xlabel('Profundidad [m]')
_ = plt.ylabel('Amplitud [1]')
```

#### Psuedo curva de tiempo a profundidad 

```python
tz = np.cumsum(np.abs(dtco))
```

```python
print('Longitud tz:',len(tz),'unidades')
```

```python
plt.figure(figsize=[13,2])
_ = plt.plot(np.linspace(0, 2000, len(tz)), tz, 'r-')
_ = plt.xlabel('Profundidad [m]')
_ = plt.ylabel('Tiempo [s]')
```

#### Interpolaci√≥n
Suponemos que la traza s√≠smica tiene un tiempo de 2 segundos. Por lo tanto, conocemos el intervalo de muestreo de los datos s√≠smicos, equivalente a 4 ms, y definimos un vector de 500 muestras equidistante, de 0 a 2000 ms. 

```python
to = 0
tf = 2000
dt = 4
t_interp = np.arange(to, tf, dt) 
```

```python
dtco_interp = np.interp(t_interp, tz, dtco)
```

```python
print('Longitud t_interp:',len(dtco_interp),'muestras')
```

```python
plt.figure(figsize=[13,2])
_ = plt.plot(dtco_interp, 'r-')
_ = plt.ylabel('Amplitud [1]')
_ = plt.xlabel('Tiempo [s]')
```

#### Seudo traza s√≠smica

```python
deltat = np.arange(0,2000,4)
sint = np.sin(0.04*deltat)
```

```python
plt.figure(figsize=[13,2])
_ = plt.plot(sint, 'b-')
_ = plt.xlabel('Tiempo [s]')
_ = plt.ylabel('Amplitud [1]')
```

```python
print('Londitud sint:',len(sint),'unidades')
```
